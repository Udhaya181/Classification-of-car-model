```python
car_devkit=Path('C:/Users/Udhaya/Desktop/archive (2)/car_devkit')
car_test=Path('C:/Users/Udhaya/Desktop/archive (2)/cars_test/cars_test')
car_train=Path('C:/Users/Udhaya/Desktop/archive (2)/cars_train/cars_train')


cars_metadata=loadmat('C:/Users/Udhaya/Desktop/archive (2)/car_devkit/devkit/cars_meta.mat')
cars_train_annos=loadmat('C:/Users/Udhaya/Desktop/archive (2)/car_devkit/devkit/cars_train_annos.mat')
cars_test_annos=loadmat('C:/Users/Udhaya/Desktop/archive (2)/car_devkit/devkit/cars_test_annos.mat')
```

```python
cars_meta=[]
for x in cars_metadata['class_names'][0]:
    cars_meta.append(x)
    
cars_classes=pd.DataFrame(cars_meta,columns=['cars_classes_exist_in_data'])
cars_classes
```

```python
fname=[[x.flatten()[0] for x in i]  for i in cars_train_annos['annotations'][0]]
column_list=['bbox_x1','bbox_y1','bbox_x2','bbox_y2','class','fname']
train_df=pd.DataFrame(fname,columns=column_list)
train_df['class']=train_df['class'] - 1 ### all values start from zero because above class_classes df index for classes started from zero hence.
train_df['fname']=[car_train/i for i in train_df['fname']]
train_df.head()
```

```python
train_df=train_df.merge(cars_classes,left_on='class',right_index=True)
train_df=train_df.sort_index()
train_df
```

```python
import os 
import zipfile

zf=zipfile.ZipFile('Stanford Cars Dataset simplified.zip',mode='w')

try:
    for i in train_df.index:
        try:
            name=train_df['cars_classes_exist_in_data'][i]
            #print(name)
            file_path=train_df['fname'][i]
            file_name=os.path.basename(train_df['fname'][i])
            short_name=name.split(" ")[0]
            zf.write(file_path,os.path.join(short_name,file_name),zipfile.ZIP_DEFLATED)
        except Exception as exc:
            print(str(exc))
            pass
finally:
    print('closing')
    zf.close()
```

```python
import os
import shutil
from pathlib import Path
from scipy.io import loadmat

# Set paths to the data directories and metadata files
car_devkit = Path('C:/Users/Udhaya/Desktop/archive (2)/car_devkit')
car_train = Path('C:/Users/Udhaya/Desktop/archive (2)/cars_train/cars_train')
cars_metadata = loadmat(car_devkit / 'devkit' / 'cars_meta.mat')
cars_train_annos = loadmat(car_devkit / 'devkit' / 'cars_train_annos.mat')

# Get the list of class names from the metadata file
class_names = [c[0] for c in cars_metadata['class_names'][0]]

# Create a subdirectory for each class
for class_name in class_names:
    os.makedirs(car_train / class_name, exist_ok=True)

# Move the images to their respective class directories
for i in range(len(cars_train_annos['annotations'][0])):
    filename = cars_train_annos['annotations'][0][i][5][0]
    class_name = class_names[cars_train_annos['annotations'][0][i][4][0].item()-1]
    src_path = car_train / filename
    dst_path = car_train / class_name / filename
    shutil.move(src_path, dst_path)
```

```python
import os
import shutil
import random

# Set the paths for the original dataset and the output folders
input_folder = "C:/Users/Udhaya/Desktop/archive (2)/cars_train/cars_train"
output_folder_1 = "C:/Users/Udhaya/Desktop/archive (2)/cars_train/cars_train (1)"
output_folder_2 = "C:/Users/Udhaya/Desktop/archive (2)/cars_train/cars_valid"

# Set the desired split ratio
split_ratio = 0.9 # 80% train and 20% test

# Create the output folders if they don't exist
if not os.path.exists(output_folder_1):
    os.makedirs(output_folder_1)
if not os.path.exists(output_folder_2):
    os.makedirs(output_folder_2)
for subfolder in os.listdir(input_folder):
    subfolder_path = os.path.join(input_folder, subfolder)
    

# Get a list of all image files in the input folder
    # Get a list of all image files in the subdirectory with the extensions .jpg, .jpeg, .png, .ppm, .bmp, .pgm, .tif, .tiff, .webp, and .gif
    files = [f for f in os.listdir(subfolder_path) if os.path.isfile(os.path.join(subfolder_path, f)) and f.lower().endswith(('.jpg', '.jpeg', '.png', '.ppm', '.bmp', '.pgm', '.tif', '.tiff', '.webp', '.gif'))]


# Shuffle the files randomly
    random.shuffle(files)

# Calculate the number of files for each output folder based on the split ratio
    split_index = int(len(files) * split_ratio)
    num_files_1 = split_index
    num_files_2 = len(files) - split_index
    subfolder_output_folder_1 = os.path.join(output_folder_1, subfolder)
    subfolder_output_folder_2 = os.path.join(output_folder_2, subfolder)
    os.makedirs(subfolder_output_folder_1, exist_ok=True)
    os.makedirs(subfolder_output_folder_2, exist_ok=True)

# Copy the files to the output folders based on the split ratio
    for i in range(num_files_1):
        shutil.copy(os.path.join(subfolder_path, files[i]), os.path.join(subfolder_output_folder_1, files[i]))
    for i in range(num_files_1, len(files)):
        shutil.copy(os.path.join(subfolder_path, files[i]), os.path.join(subfolder_output_folder_2, files[i]))
```
